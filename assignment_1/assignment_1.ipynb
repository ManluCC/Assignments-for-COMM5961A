# Method 1: Split the list of dictionaries into specific task lists and make a line chart.

```python
import pandas as pd
from matplotlib import pyplot as plt
df = pd.read_csv('monthly_hrs.csv')
df
hrs_list = df.values.tolist()
print('Monthly Hours List:','\n',hrs_list,'\n')

reading_hrs_list = []
coding_hrs_list = []
knime_hrs_list = []
week_of_month = []
weekly_hrs_reading = []
weekly_hrs_coding = []
weekly_hrs_knime = []
avg_hrs = []

for i in hrs_list:
    if i[1] == 'reading':
        reading_hrs_list.append(i)
    elif i[1] == 'coding':
        coding_hrs_list.append(i)
    elif i[1] == 'Knime exercises':
        knime_hrs_list.append(i)
print('Monthly Hours List for Each Task:','\n',reading_hrs_list)
print(coding_hrs_list)
print(knime_hrs_list,'\n')

for r in reading_hrs_list:
    week_of_month.append(r[0])
    weekly_hrs_reading.append(r[2])
for c in coding_hrs_list:
    weekly_hrs_coding.append(c[2])
for k in knime_hrs_list:
    weekly_hrs_knime.append(k[2])
print(week_of_month)
print(weekly_hrs_reading)
print(weekly_hrs_coding)
print(weekly_hrs_knime,'\n')


from cal_avg import *
for i in range(4):
    avg_hrs.append(avg(weekly_hrs_reading[i],weekly_hrs_coding[i],weekly_hrs_knime[i]))
print(avg_hrs)

plt.plot(week_of_month, weekly_hrs_reading)
plt.plot(week_of_month, weekly_hrs_coding)
plt.plot(week_of_month, weekly_hrs_knime)
plt.plot(week_of_month, avg_hrs, marker = 's',linestyle = '--')
plt.legend(['Hours spent on reading','Hours spent on coding','Hours spent on Knime exercises','average hours spent on tasks'])
plt.xlabel('Week of the month')
plt.ylabel('Hours')
plt.title('Time Spent on Learning Tasks in a Month')
plt.show()
```

# Method 2: Use Matplotlib functions to generate the line chart directly.

```python
import pandas as pd
from matplotlib import pyplot as plt
df = pd.read_csv('monthly_hrs.csv')
reading = df[df.task == 'reading']
coding = df[df.task == 'coding']
knime = df[df.task == 'Knime exercises']

avg_hrs = []
from cal_avg import *
for i in range(4):
    avg_hrs.append(avg(reading.hours.iloc[i],coding.hours.iloc[i],knime.hours.iloc[i]))
print(avg_hrs)

plt.plot(reading.week,reading.hours)
plt.plot(coding.week,coding.hours)
plt.plot(knime.week,knime.hours)
plt.plot(knime.week,avg_hrs,marker = 's',linestyle = '--')
plt.legend(['Hours spent on reading','Hours spent on coding','Hours spent on Knime exercises','average hours spent on tasks'])
plt.xlabel('Week of the month')
plt.ylabel('Hours')
plt.title('Time Spent on Learning Tasks in a Month')
plt.show()
```
